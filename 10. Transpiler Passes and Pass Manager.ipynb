# Example

**How many gates are in the quantum circuit that results from transpiling the following quantum circuit using only the pass LookaheadSwap? Assume the following coupling map: [[0, 1], [1, 2], [2, 3], [3, 4]]**

```python
%matplotlib inline
from qiskit import QuantumCircuit
from qiskit.transpiler import CouplingMap, Layout
from qiskit.transpiler.passes import BasicSwap, LookaheadSwap, StochasticSwap
from qiskit.transpiler import PassManager

coupling = [[0, 1], [1, 2], [2, 3], [3, 4]]
circuit = QuantumCircuit(5)
circuit.h(3)
circuit.cx(0, 4)
circuit.cx(3, 0)
circuit.cx(2, 4)
circuit.draw(output='mpl')
```

```python
coupling_map = CouplingMap(couplinglist=coupling)
las = LookaheadSwap(coupling_map=coupling_map)
pass_manager = PassManager(las)
```

```python
basic_circ = pass_manager.run(circuit)
```

```python
basic_circ.draw(output='mpl')
```
