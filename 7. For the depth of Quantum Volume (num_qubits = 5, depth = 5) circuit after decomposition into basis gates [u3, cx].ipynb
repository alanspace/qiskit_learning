```python
%matplotlib inline
#%config InlineBackend.figure_format = 'svg'
import numpy as np
from qiskit import QuantumCircuit, execute
from qiskit.transpiler import CouplingMap, PassManager
from qiskit.transpiler.passes import LookaheadSwap
import qiskit.circuit.library as qulib
import qiskit.quantum_info as qi
```

Specify the parameter in the quantum volume and name it QV

```python
QV = qulib.QuantumVolume(num_qubits=5, depth=5)
```

To have a shallow decomposition, we use decompose().

```python
QV_decompose = QV.decompose()
```

We now draw the circuit. 

```python
QV_decompose.draw('mpl', fold = 100)
```

From the above diagram, we see the layer of su4 elements that compose the random circuit.

Now, we decompose the circuit again with the underlaying layer of u3 and cx gates that compose the su4 blocks.

```python
QV_decompose = QV_decompose.decompose()
```

```python
QV_decompose.draw('mpl', fold = 100)
```

We now print the depth of the circuit.

```python
print(QV_decompose.depth())
```
